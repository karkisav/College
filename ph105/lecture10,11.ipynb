```python
import numpy as np
import matplotlib.pyplot as plt

Radius = 2.0
radius = 0.5
wplanet = 2.0
wmoon = 14.0
time = np.arange(0.0, 3.2, 0.02)

x0 = Radius * np.cos(wplanet * time)
y0 = Radius * np.sin(wplanet * time)

x = Radius * np.cos(wplanet * time) + radius * np.cos(wmoon * time)
y = Radius * np.sin(wplanet * time) + radius * np.sin(wmoon * time)

plt.plot(x, y, label='Moon')
plt.plot(x0, y0, label='Planet')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Motion of Planet and Moon')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Radius = 2.0
radius = 0.5
wplanet = 2.0
wmoon = 14.0
wsatellite = 20.0  # Angular velocity of the satellite orbiting the moon
time = np.arange(0.0, 3.2, 0.02)

x0 = Radius * np.cos(wplanet * time)
y0 = Radius * np.sin(wplanet * time)

x_moon = Radius * np.cos(wplanet * time) + radius * np.cos(wmoon * time)
y_moon = Radius * np.sin(wplanet * time) + radius * np.sin(wmoon * time)

r_satellite = 0.09 * np.sin(2 * np.pi * wsatellite * time)  # Adjust the function for desired spiral motion
x_satellite = x_moon + r_satellite * np.cos(wsatellite * time)  
y_satellite = y_moon + r_satellite * np.sin(wsatellite * time)  

plt.plot(x_moon, y_moon, label='Moon')
plt.plot(x0, y0, label='Planet')
plt.plot(x_satellite, y_satellite, label='Satellite')  # Plotting the satellite orbiting the moon
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Motion of Planet, Moon, and Satellite in Spirals')
plt.legend()
plt.grid()
plt.show()
```
