```python
import numpy as np
import matplotlib.pyplot as plt

Radius = 2.0
radius = 0.5
wplanet = 2.0
wmoon = 14.0
time = np.arange(0.0, 3.2, 0.02)

x0 = Radius * np.cos(wplanet * time)
y0 = Radius * np.sin(wplanet * time)

x = Radius * np.cos(wplanet * time) + radius * np.cos(wmoon * time)
y = Radius * np.sin(wplanet * time) + radius * np.sin(wmoon * time)

plt.plot(x, y, label='Moon')
plt.plot(x0, y0, label='Planet')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Motion of Planet and Moon')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0, 5.0, 0.001)
y = np.sin(30 * x) + np.sin(33 * x)
plt.ylabel("f(x) = sin(30x) + sin(33x)")
plt.xlabel("x")
plt.title("f(x) = sin(30x) + sin(33x)")
plt.plot(x, y)
plt.grid()
```
