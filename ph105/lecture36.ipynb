```python
import random
import matplotlib.pyplot as plt
```

```python
random.seed(7)
```

```python
jmax = 100
x = 0.
y = 0.
xdata,ydata = [],[]
plt.figure()
for i in range(0,jmax + 1):
    x += (random.random() - 0.5)*2
    y += (random.random() - 0.5)*2
    xdata.append(x)
    ydata.append(y)
plt.plot(xdata,ydata, 'r*-')
```
