```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-5,5, 10000)
y = pow(math.e, -x)
plt.plot(x, y)
plt.grid()
plt.title("my first graph")
plt.xlabel("x-cord")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# To generate random data for plotting
axisX = np.linspace(1, 15, 50)
axisY = np.sin(axisX)
# To create the line graph from above data
plt.plot(axisX, axisY)
# Adding title and labels for the graph
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# to show the final graph
plt.grid()
```

```python
x = np.linspace(0, 20, 1000)
y = pow(x, 4)*pow(math.e, -2*x)
z = pow(np.exp(-x)*np.sin(x**2), 2)
plt.plot(x, y, "g", label = "function 1")
plt.plot(x, z, "r", label = "function 2")
plt.title('2 Functions')
plt.xlim(0,8)
plt.xlabel("Theta")
plt.legend()
plt.grid()
```
