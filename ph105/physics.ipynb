```python
x = np.linspace(0, 20, 1000)
y = pow(x, 4)*pow(math.e, -2*x)
z = pow(np.exp(-x)*np.sin(x**2), 2)
plt.plot(x, y, "g", label = "function 1")
plt.plot(x, z, "r", label = "function 2")
plt.title('2 Functions')
plt.xlim(0,8)
plt.xlabel("Theta")
plt.legend()
plt.grid()
```

```python
import math
r_p = 4.0
r_m = 1.0
r_s = 0.125

w_p = 1.0
w_m = 13.63
w_s = 200.0

t = np.arange(0.0, 3.2, 0.02)

x_p = r_p * np.cos(w_p * t)
y_p = r_p * np.sin(w_p * t)

x_m = x_p + r_m * np.cos(w_m * t)
y_m = y_p + r_m * np.sin(w_m * t)

x_s = x_m + r_s * np.cos(w_s * t)
y_s = y_m + r_s * np.sin(w_s * t)

plt.plot(x_p, y_p, label='Planet')
plt.plot(x_m, y_m, label='Moon')
plt.plot(x_s, y_s, label='Satellite')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Motion of Planet, Moon, and Satellite')
plt.legend()
plt.grid()
plt.show()
```
