```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-5,5, 10000)
y = pow(math.e, -x)
plt.plot(x, y)
plt.grid()
plt.title("my first graph")
plt.xlabel("x-cord")
```

```python
x = np.linspace(0, 20, 1000)
y = pow(x, 4)*pow(math.e, -2*x)
z = pow(np.exp(-x)*np.sin(x**2), 2)
plt.plot(x, y, "g", label = "function 1")
plt.plot(x, z, "r", label = "function 2")
plt.title('Br2 functions')
plt 
plt.xlabel("Theta")
plt.legend()
plt.grid()
```
